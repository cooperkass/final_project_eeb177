```python
def makePeriodsList(filename):
    # this function opens the sea grass data file and extracts
    # the species name
    PeriodsList = []
    # open the file
    periods_object = open(filename,"r")
    # read the file and save contents as a list
    periods_list = periods_object.readlines()
    for line in periods_list:
        period = line.split(",")[8]
        # stick periods into PeriodsList
        PeriodsList.append(period)
    return PeriodsList
```

```python
makePeriodsList("sea_grass_no_headers.csv")
```

```python
def makeExistenceTimesList(filename):
    # this function opens the sea grass data file and extracts
    # the approximate times of existence
    ExistenceTimesList = []
    # open the file
    existence_times_object = open(filename, "r")
    # read the file and save contents as a list
    existence_times_list = existence_times_object.readlines()
    for line in existence_times_list:
        existence_time = line.split(",")[10:12]
        # stick existence_time into ExistenceTimesList
        ExistenceTimesList.append(existence_time)
    return ExistenceTimesList
```

```python
makeExistenceTimesList("sea_grass_no_headers.csv")
```

```python
def makeSpeciesList(filename):
    # this function opens the sea grass data file and extracts
    # the species name
    SpeciesList = []
    # open the file
    species_object = open(filename, "r")
    # read the file and save contents as a list
    species_list = species_object.readlines()
    for line in species_list:
        species = line.split(",")[5]
        # stick species into SpeciesList
        SpeciesList.append(species)
    return SpeciesList
print(SpeciesList)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 100.0
SpeciesList = makeSpeciesList("sea_grass_no_headers.csv")
PeriodsList = makePeriodsList("sea_grass_no_headers.csv")

species_dict = {}

for species in SpeciesList:
    species_count = SpeciesList.count(species)
    species_dict[species] = species_count
unique_species = species_dict.keys()
unique_species_occurences = list(species_dict.values())

periods_dict = {}

for period in PeriodsList:
    periods_count = PeriodsList.count(period)
    periods_dict[period] = periods_count
unique_periods = periods_dict.keys()
unique_periods_occurences = list(periods_dict.values())
    

## making charts
fig1, ax1 = plt.subplots()
ax1.pie(unique_species_occurences, labels=unique_species, autopct='%1.1f%%',
        shadow=True, startangle=0)
ax1.axis('equal')
fig = plt.gcf()
fig.set_size_inches(100,100)
plt.show()

```
